### üñáÔ∏è subplot( )ÏúºÎ°ú ÏÑúÎ∏åÌîåÎ°Ø ÏßëÌï© Ï∂îÍ∞Ä
* subplot() : ÌòÑÏû¨ figureÍ∞ùÏ≤¥Ïóê ÏÑúÎ∏åÌîåÎ°Ø ÏßëÌï© Ï∂îÍ∞Ä

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.arange(0, 10, 0.01)

fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0,0].plot(x, np.sin(x)); axes[0,1].plot(x, np.cos(x))
axes[1,0].plot(x, np.tanh(x)); axes[1,1].plot(x, np.sin(x)*np.cos(x))
```

### üñáÔ∏è plot( )
plot( ) : Ï£ºÏñ¥ÏßÑ x, y Í∞íÏùÑ ÏÑ†(lines)Í≥º Ï†ê(markers)ÏúºÎ°ú ÌëúÏãúÌï¥Ï§ÄÎã§.

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0, 10, 0.6)
y = np.sin(x)
fig, axes = plt.subplots(2, 2, figsize=(8, 5))
fig.suptitle('figure sample plots')
axes[0,0].plot(y) # ÏÑ†Í∑∏ÎûòÌîÑ, Ï†ê ÏóÜÏùå, xÏ∂ïÏùÄ range(len(y))
axes[0,1].plot(x,y,'ro:') # red(r), Ïõê(o), Ï†êÏÑ†(:)
axes[1,0].plot(x, np.cos(x), 'cs-.') # cyan(c), square(s), ÎåÄÏãúÎã∑(-.)
axes[1,1].plot([3,6], [3,5], 'b^:') # ÌååÎûë(b), ÏÑ∏Î™®(^), Ï†êÏÑ†(:)
axes[1,1].plot([4,5], [5,4], 'kx--') # Í≤ÄÏ†ï(k), X(x), ÎåÄÏãúÏÑ†(--)
plt.show()
```

### üñáÔ∏è Ï∂ïÍ≥º ÎààÍ∏à
* Ï∂ï(Axis)Í≥º ÎààÍ∏à(Tick)ÏùÄ Artist ÌÅ¥ÎûòÏä§Î•º ÏÉÅÏÜç
    * XÏ∂ïÍ≥º YÏ∂ïÏùÄ Axis ÌÅ¥ÎûòÏä§Î•º ÏÉÅÏÜç
    * XÏ∂ï ÎààÍ∏àÍ≥º YÏ∂ï ÎààÍ∏àÏùÄ Tick ÌÅ¥ÎûòÏä§Î•º ÏÉÅÏÜç
* Ï∂ïÍ≥º ÎààÍ∏àÏùò ÏßÄÏ†ï
    * Axis ÌÅ¥ÎûòÏä§Ïùò set_x???( ) Î©îÏÑúÎìúÎ•º Ïù¥Ïö© XÏ∂ïÏùò Ïä§ÌÉÄÏùºÏùÑ ÏßÄÏ†ï
    * XAxis ÌÅ¥ÎûòÏä§Ïùò set_???( ) Î©îÏÑúÎìúÎ•º Ïù¥Ïö©Ìï¥ÏÑúÎèÑ XÏ∂ïÏùò Ïä§ÌÉÄÏùºÏùÑ ÏßÄÏ†ï

```python
import numpy as np
x = np.linspace(0, 10, 100)
y = np.cos(x)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(8,5))
plt.subplots_adjust(hspace=0.4, wspace=0.3)
axes[0,0].scatter(x, y, marker=".")
axes[0,0].set(title="An Example Axes", ylabel="Y-Axis", xlabel="X-Axis")
axes[0,1].scatter(x, y, marker="^", c="r")
axes[0,1].set_xlim(0,5)
axes[0,1].set_ylim(-2,2)
axes[0,1].set_xlabel("X 0~5")
axes[0,1].set_ylabel("Y -2~2")

axes[1,0].scatter(x, y, marker=".", c="c")
axes[1,0].set(xticks=range(1,8,2),
              xticklabels=[3,100,-12,"foo"],
              yticks=[-2,0,1,2],
              yticklabels=[-200,0,100,"Max"])

axes[1,1].scatter(x, y, marker="v", c="b")
axes[1,1].set_xticks(range(1,8,2))
axes[1,1].set_xticklabels([3,100,-12,"foo"])
axes[1,1].set_yticks([-2,0,1,2])
axes[1,1].set_yticklabels([-200,0,100,"Max"])
axes[1,1].spines['top'].set_visible(False)
axes[1,1].spines['bottom'].set_position(("outward", 10))
axes[1,1].grid(True)
plt.show()
```

### üñáÔ∏è Ï∂ï Í≥µÏú†
* ÌïòÎÇòÏùò Í∑∏ÎûòÌîÑ ÏòÅÏó≠Ïóê Îëê Í∞ú Ïù¥ÏÉÅÏùò Í∑∏ÎûòÌîÑÎ•º Í∑∏Î¶¨Î©¥ÏÑú Îã§Î•∏ Ï∂ïÏùÑ ÏßÄÏ†ïÌïòÍ≥† Ïã∂ÏùÑ Í≤ΩÏö∞ twinx( ), twiny( )Î°ú ÏÉàÎ°úÏö¥ Ï∂ï Í∞ùÏ≤¥Î•º ÏÉùÏÑ±ÌïòÏó¨ Í∑∏ÎûòÌîÑÎ•º Í∑∏Î¶¨Í≥† Ï∂ïÏùÑ Îã§Î•¥Í≤å ÏßÄÏ†ïÌï† Ïàò ÏûàÎã§.
* twinx( ) Ìï®ÏàòÎäî xÏ∂ïÏùÑ Í≥µÏú†ÌïòÍ≥† yÏ∂ïÏùÑ Ïò§Î•∏Ï™ΩÏóê ÌëúÏãúÌï¥Ï£ºÎäî ÏÉàÎ°úÏö¥ Ï∂ï Í∞ùÏ≤¥Î•º Î∞òÌôòÌïúÎã§.
* twiny( ) Ìï®ÏàòÎäî yÏ∂ïÏùÑ Í≥µÏú†ÌïòÍ≥† xÏ∂ïÏùÑ ÏôºÏ™ΩÏóê ÌëúÏãúÌï¥Ï£ºÎäî ÏÉàÎ°úÏö¥ Ï∂ï Í∞ùÏ≤¥Î•º Î∞òÌôòÌïúÎã§.

```python
import numpy as np
x = np.arange(0, 10)
y1 = 0.5 * x**2
y2 = -1 * y1
```

```python
fig, ax1 = plt.subplots()
ax1.plot(x, y1, 'g^:')
ax1.set_xlabel('X data')
ax1.set_ylabel('Y1 data', color='g')

ax2 = ax1.twinx()
ax2.plot(x, y2, 'bv--')
ax2.set_ylabel('Y2 data', color='b')

ax3 = ax1.twiny()
ax3.plot(-x, y1, 'ro-.')
ax3.set_xlabel('-X data', color='r')

plt.show()
```

### üñáÔ∏è Î≤îÎ°Ä ÌëúÏãú
* Î≤îÎ°ÄÎ•º ÌëúÏãúÌïòÎäî Í∞ÄÏû• Ï¢ãÏùÄ Î∞©Î≤ïÏùÄ Í∑∏ÎûòÌîÑ Ìï®ÏàòÏóê ÏùòÌï¥ Î≤îÎ°ÄÍ∞Ä ÏûêÎèôÏúºÎ°ú Í∑∏Î†§ÏßÄÍ≤å ÌïòÎäî Î∞©Î≤ï
* legend( ) Ìï®ÏàòÎäî plot( ) Ìï®ÏàòÍ∞Ä label ÏÜçÏÑ±ÏùÑ Í∞ÄÏßà Í≤ΩÏö∞ Ìï®ÏàòÏùò Ìò∏Ï∂úÎßåÏúºÎ°ú Î≤îÎ°ÄÎ•º ÌëúÏãúÌïúÎã§.
* plot( ) Ìï®ÏàòÏóê label ÏÜçÏÑ±ÏùÑ ÏßÄÏ†ïÌïòÏßÄ ÏïäÏùÑ Í≤ΩÏö∞ legend( ) Ìï®ÏàòÏùò labels ÌååÎùºÎØ∏ÌÑ∞Î•º Ïù¥Ïö©Ìï¥ÏÑú Î≤îÎ°ÄÏùò Î†àÏù¥Î∏î(Ïù¥Î¶Ñ)ÏùÑ ÏßÄÏ†ïÌïúÎã§.

```python
import numpy as np
x = np.arange(0, 10)
y = 0.5 * x**2
```

```python
plt.style.use('default')
fig, ax1 = plt.subplots(figsize=(8,4))
ax1.plot(x, y, 'g^:', label='GREEN')
ax1.plot(x, -y, 'bv-.', label='BLUE')
ax1.legend() # label ÏÜçÏÑ± Í∞íÏùÑ ÎØ∏Î¶¨ ÏßÄÏ†ïÌñàÎã§Î©¥ legend() Ìï®Ïàò Ìò∏Ï∂úÎßåÏúºÎ°ú Î≤îÎ°ÄÎ•º ÎÇòÌÉÄÎÇº Ïàò ÏûàÎã§.
plt.show()
```

```python
# plot() Ìï®ÏàòÏóê label ÏÜçÏÑ±ÏùÑ ÏßÄÏ†ïÌïòÏßÄ ÏïäÏïòÎã§Î©¥ legend() Ìï®ÏàòÏùò labels ÌååÎùºÎØ∏ÌÑ∞Î•º Ïù¥Ïö©ÌïòÏó¨ Î≤îÎ°ÄÏùò Î†àÏù¥Î∏î(Ïù¥Î¶Ñ)ÏùÑ ÏßÄÏ†ïÌï† Ïàò ÏûàÎã§.
plt.style.use('default')
fig, ax1 = plt.subplots(figsize=(8,4))
ax1.plot(x, y, 'g^:')
ax1.plot(x, -y, 'bv-.')
ax1.legend(labels=["GREEN", "BLUE"])
plt.show()
```
