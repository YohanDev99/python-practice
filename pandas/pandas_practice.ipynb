## ‚úÖ iris Îç∞Ïù¥ÌÑ∞ÏÖã

* Î∂ìÍΩÉ(iris)Ïùò 3Í∞ÄÏßÄ Ï¢Ö(setosa(ÏÑ∏ÌÜ†ÏÇ¨), versicolor(Î≤ÑÏãúÏª¨Îü¨), virginica(Î≤ÑÏßÄÎãàÏπ¥)) Î≥ÑÎ°ú Í∞ÅÍ∞Å 50Í∞ú Îç∞Ïù¥ÌÑ∞Ïùò ÍΩÉÎ∞õÏπ®Ïùò Í∏∏Ïù¥ÏôÄ ÎÑàÎπÑ, ÍΩÉÏûéÏùò Í∏∏Ïù¥ÏôÄ ÎÑàÎπÑÎ•º ÏÑºÌã∞ÎØ∏ÌÑ∞ Îã®ÏúÑÎ°ú Ï∏°Ï†ïÌïòÏó¨ Ï†ïÎ¶¨Ìïú Îç∞Ïù¥ÌÑ∞ÏÖã

### üñáÔ∏è Scikit-learn Ìå®ÌÇ§ÏßÄÏùò iris Îç∞Ïù¥ÌÑ∞ÏÖã
* Ïã∏Ïù¥ÌÇ∑Îü∞Ïùò load Í≥ÑÏó¥ Ìï®ÏàòÎì§ Ï§ëÏóêÏÑú load_iris()Î•º Ïù¥Ïö©ÌïúÎã§.
* sklearnÏùò iris Îç∞Ïù¥ÌÑ∞Îäî dict ÌòïÏãùÏù¥Î©∞ data, target, target_name, DESCR, feature_names Îì±ÏùÑ ÌÇ§(key)Î•º Í∞ÄÏßÑÎã§.

```python
import pandas as pd
from sklearn import datasets
```

```python
iris_dic = datasets.load_iris()
iris_dic
```

```python
x = pd.DataFrame(iris_dic['data'], columns=iris_dic['feature_names'])
y = pd.DataFrame(iris_dic['target_names'][iris_dic['target']], columns=['species'])
```

```python
iris = pd.concat([x,y], axis=1)
iris.head()
```

```python
# ÏÉòÌîå Îç∞Ïù¥ÌÑ∞
import statsmodels.api as sm
airquality_data = sm.datasets.get_rdataset("airquality")
airquality = airquality_data.data
airquality.head()
```

### üñáÔ∏è melt()Î•º Ïù¥Ïö©Ìïú Ïñ∏ÌîºÎ≤óÌåÖ

### üñáÔ∏è pivot_table()ÏùÑ Ïù¥Ïö©Ìïú ÌîºÎ≤óÌåÖ
