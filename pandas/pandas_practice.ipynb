## ✅ 판다스 소개
*   판다스 : 시리즈(Series, 1차원)와 데이터프레임(DataFrame, 2차원)을 제공
*   데이터프레임 : 데이터프레임은 테이블 형식이고 이종모음들로 구조화된 데이터
*   시리즈 : 데이터프레임에서 열(Column) 하나를 의미, 주로 시계열 데이터를 표현한다.
*   판다스는 재무, 통계, 사회 과학 등 다양한 분야의 데이터를 처리하기 위해서 사용한다.
*   부분집합 조회 열 추가 및 제거, 병합, 데이터 구조 변경 등 데이터 전처리를 위한 기능을 제공한다.

### 🖇️ DataFrame
* 행(Row)
    * 1개의 행은 각각 다른 데이터를 갖는 튜플
    * 행은 이름과 인덱스(위치)를 갖는다.
    * 행의 이름은 인덱스와 같을 경우가 많다.
* 열(Column)
    * 열 내의 모든 데이터는 같은 타입이다.
    * 열의 이름과 인덱스(위치)를 갖는다.

## ✅ 딕셔너리를 이용해서 데이터프레임 만들기

## ✅ 리스트를 이용해서 데이터프레임 만들기
*   딕셔너리를 이용해 데이터프레임을 생성하는 경우가 아니라면 일반적으로 열 이름을 따로 지정해줘야 한다.

```python
# 리스트를 딕셔너리의 값으로 지정하여 데이터프레임을 생성할 수 있다.
df = pd.DataFrame({'col1' : L1, 'col2' : L2})
df
```

## ✅ 데이터프레임 병합과 연결

### 🖇️ merge()를 이용한 데이터프레임 병합
* merge() 함수는 두 개 데이터프레임의 공통 열 또는 공통 인덱스를 기준으로 두 개의 데이터프레임을 하나로 합친다.

```python
# how속성
df1 = pd.DataFrame({'key':['a', 'b', 'c', 'f'], 'c1':[1,2,3,5]})
df2 = pd.DataFrame({'key':['a', 'b', 'd', 'f'], 'c2':[5,6,7,8]})
```

```python
df1.merge(df2)
```

```python
df1.merge(df2, how="left")
```

```python
df1.merge(df2, how="right")
```

```python
df1.merge(df2, how="outer")
```

### 🖇️ left_on, right_on 속성
* 키가 되는 기준열의 이름이 두 데이터프레임에서 다르다면 left_on과 right_on 인수를 사용하여 기준이 되는 열을 명시해야 한다.

```python
df3 = pd.DataFrame({'key3':['a', 'b', 'c', 'f'], 'c1':[1, 2, 3, 5]})
df4 = pd.DataFrame({'key4':['a', 'b', 'd', 'f'], 'c2':[5, 6, 7, 8]})
```

```python
df3.merge(df4, left_on="key3", right_on="key4")
```

### 🖇️ concat( )을 이용한 데이터프레임 연결
* concat( ) 함수는 두 데이터프레임을 연결한다.

```python
# axis 속성
df1 = pd.DataFrame({'c1': [1, 2, 3, 4],
                    'c2': [5, 6, 7, 8]})
df2 = pd.DataFrame({'c3': ['a', 'b', 'c', 'd'],
                    'c4': [1.2, 3.4, 5.5, 7.6]})
```

```python
# concat()은 axis 속성에 따라 axis=0이면 데이터프레임을 위에서 아래로 연결하며, axis=1이면 왼쪽에서 오른쪽으로 연결한다.
pd.concat([df1, df2], axis=1)
```

```python
# axis=0일 경우 주의사항
df1 = pd.DataFrame({'c1': [1, 2], 'c2': [3, 4]})
df2 = pd.DataFrame({'c1': [5, 6], 'c4': [7, 8]})
```

```python
# axis=0일 경우 같은 이름의 열 이름이 있을 경우 위에서 아래로 데이터를 쌓아 연결해 주지만
# 열의 이름이 다를 경우 없는 곳의 데이터가 NaN 값이 된다.
pd.concat([df1, df2], axis=0, sort=False)
```

### 🖇️ concat()과 reset_index()
* 연결할 때 내부적으로 인덱스를 사용하므로 두 데이터프레임의 인덱스가 다를 경우에는 reset_index()를 이용해서 인덱스를 초기화 해줘야 한다.

```python
df1 = pd.DataFrame({'c1': [1, 2, 3, 4],
                    'c2': [5, 6, 7, 8]}, index=[0, 2, 4, 6])
df2 = pd.DataFrame({'c3': ['a', 'b', 'c', 'd'],
                    'c4': [1.2, 3.4, 5.5, 7.6]}, index=[0, 1, 2, 3])
```

```python
pd.concat([df1, df2], axis=1, sort=False)
```

```python
# reset_index() 매우 중요 # drop은 true로 # 51페이지 참고
df1.reset_index(drop=True, inplace=True)
pd.concat([df1, df2], axis=1)
```

## ✅ 데이터 그룹화 및 집계

### 🖇️ groupby
* groupby()는 데이터를 구분할 수 있는 열(column)의 값들을 이용하여 데이터를 여러 기준에 의해 구분하여 그룹화 한 후 기초 통계 함수 등을 적용할 수 있도록 한다.

### 🖇️ 단일 열로 그룹화
* groupby 함수의 인수로 그룹화 할 열을 지정한다.

```python
import seaborn as sns
iris = sns.load_dataset("iris")
```

```python
iris_grouped = iris.groupby(by=iris.species)
iris_grouped
```

```python
iris_grouped.mean()
```

### 🖇️ 다중 열로 그룹화
* groupby 함수의 인수로 그룹화 할 열을 리스트 형식으로 지정한다.

```python
import numpy as np
iris["num"] = np.ravel([[i]*25 for i in range(6)]) # 2차원을 1차원으로 풀어주는 구문
iris.head()
```

```python
iris_grouped2 = iris.groupby(by=[iris.species, iris.num])
iris_grouped2.mean()
```

### 🖇️ 그룹간 반복 처리
* 그룹화 된 데이터에서 그룹의 타입과 그룹 객체를 반복문을 이용해 처리가 가능하다.

```python
# 그룹화 된 데이터에서는 loc, iloc, 인덱싱이 불가능하다.
# 서브그룹을 만들어서 사용할 것?
for t, group in iris_grouped:
    print(t, type(group))
```

### 🖇️ 데이터프레임 그룹 인덱싱
* 데이터프레임 그룹에서의 인덱싱은 take() 함수를 이용한다.

```python
# 0번째 그룹, 1번째 그룹의 의미가 아니라 각 그룹에서 0번째, 1번째를 의미한다.
iris_grouped.take([0,1] )
```

## ✅ 데이터 구조 변경

### 🖇️ 와이드 포맷과 롱 포맷
* 와이드 포맷(wide format)
    * 가로로 긴 형식의 데이터 구조
    * 열 단위 데이터 구조
    * 피벗 테이블(pivot table)
* 롱 포맷(long format)
    * 세로로 긴 형식의 데이터 구조
    * 행 단위 데이터 구조
    * 언피벗 테이블(unpivot table)

```python
# 샘플 데이터
import statsmodels.api as sm
airquality_data = sm.datasets.get_rdataset("airquality")
airquality = airquality_data.data
airquality.head()
```

### 🖇️ melt()를 이용한 언피벗팅
* melt() 함수는 와이드 포맷 데이터를 롱 포맷 데이터로 데이터 구조를 바꿔준다.

```python
import pandas as pd
airquality_melted = pd.melt(airquality, id_vars=["Month", "Day"],
                            value_vars=["Ozone", "Solar.R", "Wind", "Temp"])
```

### 🖇️ pivot_table()을 이용한 피벗팅
* pivot_table() 함수는 피벗 테이블을 만들어준다.

```python
airquality2 = airquality_melted.pivot_table(index=["Month", "Day"],
                                            columns=["variable"], values=["value"])
```

```python
airquality2.head()
```

```python
# reset_index를 사용할 때 drop 속성을 True로 지정하면 기존 인덱스를 지워버리고 새로운 인덱스 번호를 부여하고
# drop 속성을 False로 지정하면 기존의 인덱스를 데이터프레임의 새로운 열에 추가하고 새로운 인덱스 번호를 부여한다.
airquality2 = airquality2.reset_index(level=["Month", "Day"], col_level=1)
airquality2.columns = airquality2.columns.droplevel(level=0)
```

```python
airquality2.head()
```

## ✅ 데이터프레임에 함수 적용하기

### 🖇️ apply()
* apply() 함수는 데이터프레임 또는 시리즈의 각 요소별로 특정 함수를 적용시킨다.

```python
import seaborn as sns
iris_df = sns.load_dataset("iris")
```

```python
import numpy as np
iris_df.iloc[:,:-1].apply(np.round).head()
```

```python
iris_df.iloc[:,:-1].apply(np.sum)
```

* 시리즈와 람다식의 사용

```python
iris_df2 = sns.load_dataset("iris")
```

```python
import numpy as np
iris_df2 = iris_df2.iloc[:,:-1]
```

```python
iris_avg = iris_df.iloc[:,:-1].apply(np.average)
iris_avg
```

```python
iris_df2.apply(lambda X : X-iris_avg, axis=1).head()
```

* 함수가 리스트를 반환할 경우


```python
iris_df2.apply(lambda X : list(X-iris_avg), axis=1).head()
```

```python
# result_type='broadcast' 매개변수를 설정하면 원본 데이터의 열 이름과 구조를 그대로 유지한다.
iris_df2.apply(lambda X : list(X-iris_avg), axis=1, result_type='broadcast').head()
```

### 🖇️ applymap()
* applymap() 함수는 각 요소(element)별로 작동한다.

```python
iris_df2.applymap(lambda x : x**2).head()
```

### 🖇️ map()
* map() 함수는 시리즈의 각 요소(element)별로 작동한다.
* map() 함수는 시리즈(Series)타입에서만 사용할 수 있다.

```python
iris_df.sepal_length.map(np.round)
```

```python
iris_df.sepal_length.map(np.sum)
```

* None 값을 포함할 경우 na_action 매개변수의 동작

```python
import pandas as pd
s = pd.Series([1, 2, 3, None])
```

```python
s.map(lambda x : (x, x**2))
```

```python
s.map(lambda x : (x, x**2), na_action='ignore')
```

### 🖇️ 딕셔너리 사용
* 시리즈에 딕셔너리를 적용하면 딕셔너리의 키별로 시리즈의 값에 적용된다.

```python
x = pd.Series(["Hello", "Python", "World"])
x.map({"Hello":'A', "World":'B', "Python":'C'})
```

### 🖇️ 시리즈 사용
* 시리즈에 시리즈를 적용하면 원본 시리즈의 값에 적용할 시리즈(arg)의 인덱스 별로 적용된다.

```python
y = pd.Series(['foo', 'bar', 'baz'],
              index=['Hello', 'Python', 'World'])
```

```python
x.map(y)
```

### 🖇️ na_action
* 시리즈가 NaN 값을 포함할 경우 na_action 인자의 값에 따라 결과는 달라진다.

```python
s = pd.Series([1, 2, 3, None])
s
```

```python
s.map(lambda x : (x, x**2))
```

```python
s.map(lambda x : (x, x**2), na_action='ignore')
```

## ✅ 결측치 처리 및 변경하기

```python
import seaborn as sns
iris_x = sns.load_dataset("iris")
```

```python
import random
random.seed(1)
for col in range(4) :
    iris_x.iloc[[random.sample(range(len(iris)), 20)], col] = float('nan')
```

```python
iris_x.head(15)
```

### 🖇️ dropna
* dropna() 함수는 결측치(누락된 값, NA, NaN)를 포함한 행 또는 열을 제거해준다.

```python
iris_x.dropna().head()
```

### 🖇️ fillna
* fillna() 함수는 결측치(NA, NaN)를 주어진 방법으로 채운다.

### 🖇️ replace
* replace() 함수는 to_replace 속성에 주어진 값을 value 값으로 바꾼다.
* 대체될 값을 찾는 방법을 숫자(numeric), 문자(str) 또는 정규식(regex)으로 지정할 수 있다.

```python
# 상수값으로 대체
iris_x.replace(float('nan'), 10).head()
```

```python
# 채울 방법 지정
iris_x.sepal_length.replace([5, 4.6], method="bfill").head(10)
```

```python
# 정규표현식을 이용
iris.replace("r^se[a-z]*", "set", regex=True).head(10)
```

### 🖇️ where
* where()는 하나 이상의 조건에 대한 데이터프레임을 확인하고 그에 따라 결과를 반환한다.

### 🖇️ astype
